```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parámetros
N = 1000  # Puntos de la señal
tiempo = np.linspace(0, 1, N)  # Vector de tiempo de 0 a 1 segundo

# Generación de senoidales
seno1 = np.sin(1 * 2 * np.pi * tiempo)  # 1 ciclo, amplitud 1 (digamos, 1 oscilación en 1 segundo)
seno2 = np.sin(2 * 2 * np.pi * tiempo)  # 2 ciclos, amplitud 1
seno10 = 2 * np.sin(10 * 2 * np.pi * tiempo)  # 10 ciclos, amplitud 2
seno20 = 1.5 * np.sin(20 * 2 * np.pi * tiempo)  # 20 ciclos, amplitud 1.5

# Nivel DC (constante en 1)
niveldc = np.ones(N)

# Señal combinada
seno121020dc = seno1 + seno2 + seno10 + seno20 + niveldc

# Gráficos
plt.figure(figsize=(10, 6))

plt.subplot(5, 1, 1)
plt.plot(tiempo, seno1)
plt.title("Señal con 1 ciclo")

plt.subplot(5, 1, 2)
plt.plot(tiempo, seno2)
plt.title("Señal con 2 ciclos")

plt.subplot(5, 1, 3)
plt.plot(tiempo, seno10)
plt.title("Señal con 10 ciclos")

plt.subplot(5, 1, 4)
plt.plot(tiempo, seno20)
plt.title("Señal con 20 ciclos")

plt.subplot(5, 1, 5)
plt.plot(tiempo, seno121020dc)
plt.title("Suma de todas las señales + Nivel DC")

plt.tight_layout()
plt.show()
```

```python
Fs = 1000  # Frecuencia de muestreo (Hz)

# Aplicamos la Transformada de Fourier
frecuencias = np.fft.fftfreq(N, 1/Fs)  # Eje de frecuencias
espectro = np.fft.fft(seno121020dc)  # Transformada de Fourier

# Grafico del espectro de frecuencia
plt.figure(figsize=(6, 4))
plt.xlim(-1,50)
plt.plot(frecuencias[:N//2], np.abs(espectro[:N//2]))  # Solo la mitad positiva
plt.title("Espectro de Frecuencia (Transformada de Fourier)")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.show()
```
